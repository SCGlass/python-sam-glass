## Example - birds and oil

There is an oil leakage causing the bird population in an island to 1/2 each year. From start there were 
80000 birds, how many years does it take for it to have 1/10 remaining?

```python
birds, years = 80000, 0 #tuple unpacking

limit = birds/10

while birds >= limit:
    birds /=2
    years += 1
    print(f"Year {years}: number of birds {birds:.0f} left") #:.0f so it rounds decimals

 
```

## Example - Milk
- 1L of milk has 1 500 000 bacteria in room temperature
- Bacterias increases with 50% each hour in room temp
- Milk gets soured after 10 000 000 bacterias
- How many hours does it take til milk sours (whole hours)


```python
import matplotlib.pyplot as plt 

bacterias = 1.5e6 # 1.5 * 10^6
factor = 1.5
hours = 0
sour = 1e7 #10^7

bacterias_list = [bacterias]

print(bacterias_list)

while bacterias < sour:
    bacterias *= factor
    hours += 1
    bacterias_list.append(bacterias) #add element into the list

print(bacterias_list)
print(f"Takes {hours}h for milk to get soured")

# creates a range object with numbers: 0,1,2 .... , hours
hours_list = range(0, hours+1)

plt.plot(hours_list,bacterias_list)
plt.grid()
plt.xlabel("Hours")
plt.ylabel("Number of bacterias")
plt.title("Bacteria in milk in room temparature")
```
