# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# array from -10 to 10 with 100 numbers equally spaced
x = np.linspace(-10,10, 20)

f = lambda x: np.cos(x)
# equivalent
# def f(x):
#   return np.cos(x)
```

```python
x, f(x)
```

## MATLAB style interface

```python
plt.plot(x, f(x), 'o--')
```

```python
x = np.linspace(-10,10, 200)

plt.figure(figsize=(8, 4), dpi = 200) # changes the size of the graph
plt.plot(x, f(x)) # plots points
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
```

## OOP approach

```python
# instantiated fig, ax from figure and axes
fig, ax = plt.figure(figsize = (8,4)), plt.axes()
ax.plot(x, f(x)) # ax object has a plot method
ax.set(xlabel = "x", ylabel="y", title = "$\cos{x}", xlim = (-5, 5), ylim = (-1.2, 1.2))
fig.savefig("../assets/cos(x).png", dpi=200)
```

## Subplots

```python
# MATLAB style
plt.figure(figsize=(16,5))
plt.subplot(1,2,1) # (#rows, #colums, pane1 number)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")

plt.subplot(1,2,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trigonometric funtions", fontweight = "bold")

```

```python
fig, ax = plt.subplots(1,2, figsize=(16,5)

ax[0].plot(x,f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)")
fig.suptititle("Big Title", fontsize = 25)

# If we would have 2x2 -> ex 3rd ax: ax [1][0]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

data_normal = np.random.randn(1000)

ax[0].hist(data_normal, bins=20, alpha=.7)
ax[0].set(title="Histogram of normally distributed data",
          xlabel="x", ylabel="y")


size = 1000 * np.random.randn(100)
scatter = ax[1].scatter(x=np.random.randn(100), y=np.random.randn(100), s=size, alpha=0.3, cmap="viridis", c=np.random.randn(100))

ax[1].set(title="Bubble chart", xlabel = "x", ylabel = "y");

fig.colorbar(scatter)
fig.set_facecolor("lightblue")
```
